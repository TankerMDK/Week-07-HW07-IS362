This notebook loads a movie ratings dataset into pandas, calculates average ratings by user and by movie, creates a normalized ratings dataframe, and explains the advantages and disadvantages of using normalized ratings instead of actual ratings.

The dataset includes five users rating six recent popular movies on a scale from 1 to 5. Blank cells indicate that the user has not seen that movie.

## 1. Loading the Ratings Dataset

The first step is to load the movie ratings data from a CSV file into a pandas dataframe. This allows the dataset to be inspected and analyzed.

```python
import pandas as pd

ratings = pd.read_csv("movieRatings.csv")

print(ratings)
```

## 2. Average Ratings for Each Movie and Each User

Once the data is loaded, the next step is to calculate the average rating for each movie and the average rating given by each user.

```python
movie_avg = ratings.mean(numeric_only=True)

print("Average Rating per Movie")
print(movie_avg)
```

```python
user_avg = ratings.set_index("User").mean(axis=1)

print("Average Rating per User")
print(user_avg)
```

## 3. Creating Normalized Ratings

Different users may rate movies differently. Some users may give mostly high scores while others may rate more strictly.  
To account for this, the ratings can be normalized so that each user's ratings fall between 0 and 1.

This allows comparisons between users while preserving their relative preferences.

### Preparing the Dataset for Normalization

Before performing normalization, the dataset is restructured so that the user names become the dataframe index.  
This makes it easier to perform calculations across each user's ratings.

```python
ratings_indexed = ratings.set_index("User")

normalized = (ratings_indexed - ratings_indexed.min(axis=1).values.reshape(-1,1)) / \
    (ratings_indexed.max(axis=1).values.reshape(-1,1) - ratings_indexed.min(axis=1).values.reshape(-1,1))

print("Normalized Ratings")
print(normalized)
```

### Average Normalized Rating per Movie

After normalization, the average rating for each movie can be recalculated.  
This allows us to compare how movies perform once user rating styles have been adjusted.

```python
movie_avg_norm = normalized.mean()

print("Average Normalized Rating per Movie")
print(movie_avg_norm)
```

### Average Normalized Rating per User

The average normalized rating for each user is also calculated.  
This helps confirm that the normalization process has adjusted ratings while preserving each user's relative preferences.

```python
user_avg_norm = normalized.mean(axis=1)

print("Average Normalized Rating per User")
print(user_avg_norm)
```

## 5. Conclusion

Normalized ratings help adjust for differences in how users rate movies. Some users tend to rate movies more generously, while others may give lower scores overall. By normalizing ratings, each user's preferences are scaled relative to their own rating range.

One advantage of normalized ratings is that it allows fairer comparisons between users. A user who generally rates movies lower will not unfairly influence the overall results compared to someone who rates everything highly.

A disadvantage is that normalization removes some of the absolute meaning of the ratings. A movie that received a very high rating from everyone might appear less extreme once the ratings are rescaled.

## 5. Extra Credit – Standardized Ratings

In addition to normalization, ratings can also be standardized using z-scores. Standardization shows how far each rating is from that user's average rating, measured in standard deviations.

This can be useful for identifying ratings that are unusually high or low relative to a user's typical pattern.

### Calculating Standardized Ratings

Standardization converts ratings into z-scores by subtracting the user's mean rating and dividing by the standard deviation.  
This transformation shows how far each rating is from the user's average rating.

```python
standardized = (ratings_indexed - ratings_indexed.mean(axis=1).values.reshape(-1, 1)) / (
    ratings_indexed.std(axis=1).values.reshape(-1, 1)
)

print("Standardized Ratings")
print(standardized)
```

```python
movie_avg_std = standardized.mean()

print("Average Standardized Rating per Movie")
print(movie_avg_std)
```

```python
user_avg_std = standardized.mean(axis=1)

print("Average Standardized Rating per User")
print(user_avg_std)
```

### Extra Credit Conclusion

Standardized ratings convert each rating into a value that reflects how far it is from a user's average rating. Positive values indicate that a movie was rated higher than the user's typical rating, while negative values indicate that the movie was rated lower than their usual rating. Values close to zero represent ratings that are near the user's average.

This approach helps highlight each user's relative preferences rather than relying on the raw scores alone. For example, a movie with a positive standardized value was liked more than that user's typical rating behavior. This can be useful when comparing preferences between users who may rate movies differently overall.
